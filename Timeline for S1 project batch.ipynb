```python
import os
import numpy as np
import pandas as pd
import pyabf
import matplotlib.pyplot as plt

data_path = r'/Volumes/Behaviour/2P Digitizer Data/'
files = [abf for abf in os.listdir(data_path)]

df = pd.DataFrame()

for file in files:
    abf = pyabf.ABF(data_path+file)
    print('processing ', file)
    # process the 2P trace
    abf.setSweep(sweepNumber=0, channel=0)
    y_2P = (abf.sweepY > 3) * 1
    y_2P = ((y_2P[1:] - y_2P[:-1]) == 1) * 1
    y_2P_trig = np.where(y_2P == 1)[0]+1
    x_2P = abf.sweepX
    x_2P_trig = x_2P[y_2P_trig]
    
    # process the arduino trace
    abf.setSweep(sweepNumber=0, channel=2)
    y_ard = (abf.sweepY > 3) * 1
    y_ard = ((y_ard[1:] - y_ard[:-1]) == 1) * 1
    y_ard_trig = np.where(y_ard == 1)[0]+1
    x_ard = abf.sweepX
    x_ard_trig = x_ard[y_ard_trig]
       
    # find all session start times
    session_trig = []
    session_trig.append(x_ard_trig[0])
    for i in range(x_ard_trig.shape[0]-1):
        if x_ard_trig[i+1]-x_ard_trig[i] > 30:
            session_trig.append(x_ard_trig[i+1])
    print("Found", len(session_trig), "sessions in ", file)
    session_trig = np.array(session_trig)
   
    # in the 2P image stack, find the number of the first image of each session
    sessionframe_2P = []
    for s in session_trig:
        sessionframe_2P.append(((np.where(np.round(x_2P_trig,1) == np.round(s,1))[0][0]+1)*2)) #this uses 2 frames before first trigger and then one trigger every 2 frames
        
    print("Found", len(sessionframe_2P), "first 2P images")
    #sessionframe_2P.append(((len(x_2P_trig)-1)*2)-1) # add the last frame of the recording
    #sessionframe_2P.append((sessionframe_2P[-1] - sessionframe_2P[-2])/6.1)
    sessionframe_2P = np.array(sessionframe_2P)
    sf = pd.DataFrame(sessionframe_2P,columns=[file]) 

    df = pd.concat([df, sf], axis=1)
    
df.to_csv(os.path.join(data_path, '2P_stimframes_Final.csv'))
```
