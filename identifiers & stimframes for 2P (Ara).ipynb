```python

import os
import imageio
import cv2
import math
import time
import pandas as pd
import numpy as np
import seaborn as sns
from tifffile import imsave
import matplotlib.pyplot as plt
from os.path import dirname, realpath

data_path = r'F:\S1 2P Migrated Behaviour'

directories = [x[0] for x in os.walk(data_path) if os.path.basename(x[0]) == 'faceCam' and "video_files" not in x[0]]
if len(directories) > 0:
    identifier_df = pd.DataFrame()
    for count, directory in enumerate(directories):

        #get trial information
        mouse = os.path.basename(dirname(dirname(dirname(dirname(realpath(directory))))))
        stimset = os.path.basename(dirname(dirname(dirname(realpath(directory)))))
        trial = os.path.basename(dirname(dirname(realpath(directory))))
        uniqueID = os.path.basename(dirname(realpath(directory)))
        identifier = str(mouse+'_'+stimset+'_'+trial+'_'+uniqueID)
        print(count, identifier)
        identifier_df.loc[count,'directory'] = directory
        identifier_df.loc[count,'identifier'] = identifier       

    identifier_df.to_csv(os.path.join(data_path, 'identifiers.csv'))

    print('Made identifiers')

    mice = []
    sessions = []
    stimsets = []
    uniqueIDs = []
    for i in identifier_df['identifier']:
        entries = i.split('_')
        mice.append(entries[0])
        sessions.append(entries[1])
        stimsets.append(entries[3])
        uniqueIDs.append(entries[6])

    mice = pd.DataFrame({'mice':list(set(mice))})
    sessions = pd.DataFrame({'sessions':list(set(sessions))})
    stimsets = pd.DataFrame({'stimsets':list(set(stimsets))})
    uniqueIDs = pd.DataFrame({'uniqueIDs':list(set(uniqueIDs))})
    key_terms = pd.concat([mice,sessions,stimsets,uniqueIDs],axis=1)
    key_terms.to_csv(os.path.join(data_path, 'key_terms.csv'))

    print('Got key terms')
    print('')

else:
    print('Error: no files found')
```

```python
manual_timings_path = os.path.join(data_path, 'manual_timings_2P.csv')

directories = [x[0] for x in os.walk(data_path) if os.path.basename(x[0]) == 'faceCam' and "video_files" not in x[0]]
if len(directories) > 0:
    manual_timings = pd.read_csv(manual_timings_path, header=[0])
    identifiers = pd.read_csv(os.path.join(data_path, 'identifiers.csv'),index_col=0)

    #this fixes issue of dates when csv is saved with excel
    for i in range(manual_timings['date'].size):
        manual_timings.loc[i, 'date'] =str(manual_timings.loc[i, 'date'][:2]) + str(manual_timings.loc[i, 'date'][3:6].upper())+ str('2021')
        #print(manual_timings.loc[i, 'date'])

    for i in range(manual_timings.shape[0]):
        #manual_timings_label = manual_timings.loc[i,'mouse']+'_'+manual_timings.loc[i,'condition']+'_'+manual_timings.loc[i,'date'].upper()+'_'+manual_timings.loc[i,'stimuli']+'_'+str(int(manual_timings.loc[i,'uniqueID']))
        #manual_timings_label = manual_timings.loc[i,'mouse']+'_'+manual_timings.loc[i,'session']+'_'+manual_timings.loc[i,'date'].upper()+'_'+manual_timings.loc[i,'stimuli']+'_'+str(int(manual_timings.loc[i,'uniqueID']))
        manual_timings_label = manual_timings.loc[i,'mouse']+'_'+manual_timings.loc[i,'session']+'_'+manual_timings.loc[i,'date'].upper()+'_'+str(manual_timings.loc[i,'stimuli'])+'_'+str(int(manual_timings.loc[i,'uniqueID']))
        manual_timings.loc[i,'short_identifier'] = manual_timings_label

    stimframes_df = pd.DataFrame()
    for count, directory in enumerate(directories):

        mouse = os.path.basename(dirname(dirname(dirname(dirname(realpath(directory))))))
        stimset = os.path.basename(dirname(dirname(dirname(realpath(directory)))))
        uniqueID = os.path.basename(dirname(realpath(directory)))
        identifier = identifiers.loc[identifiers['directory'].str.contains(mouse)&identifiers['directory'].str.contains(stimset)&identifiers['directory'].str.contains(uniqueID)]['identifier'].values[0]
        #print('identifier: ', identifier)
        
        #get all tif files
        image_list = [tif for tif in os.listdir(directory) if tif.endswith('.tif')]
        image_list = sorted(image_list, key=lambda name: int(name[-8:-4]))

        #check for baseline frames that were inserted by Arduino upload
        im1 = []
        for frame, image in enumerate(image_list[:100]): 
            image = imageio.imread(os.path.join(directory, image))
            im1.append(image)

        im1_motion = []
        for frame in range(1, len(im1)):
            im1_motion.append((cv2.absdiff(im1[frame], im1[frame-1])).mean())
        im1_motion = np.array(im1_motion)
        if (im1_motion>3).sum() > 0:
            inserted_frames = 1+((im1_motion>3).argmax())
            print(inserted_frames, 'inserted frames detected')
        if (im1_motion>3).sum() > 1:
            print('More than one merge detected')
        else:
            inserted_frames = 0

        short_identifier = identifier.split('_')
        short_identifier = short_identifier[0]+'_'+short_identifier[1]+'_'+short_identifier[2]+'_'+short_identifier[3]+'_'+short_identifier[6]
        #print('short_identifier: ', short_identifier)
        
        try: 
            offset = manual_timings[manual_timings['short_identifier'].str.contains(short_identifier)]['offset'].values[0]
        except:
            print('Could not find', short_identifier, 'in manual timings')
            offset = 1203

        stimframes_df.loc[count,'identifier'] = identifier
        stimframes_df.loc[count,'inserted frames'] = inserted_frames

        stimtype = identifier.split('_')[4]
        autotypes = ['air-puff','hot','warm','tone','off-target','no stim']
            
        if offset == 'DO NOT USE':
            sframe = 'DO NOT USE'
        elif stimtype in autotypes:
            sframe = 1203 + int(inserted_frames)
        elif math.isnan(float(offset)):
            sframe = 1203 + int(inserted_frames)
        else:
            sframe = int(offset) + int(inserted_frames)

        stimframes_df.loc[count,'sframe'] = sframe

        print(count, identifier)
        print('inserted frames =', inserted_frames)
        print('sframe =', sframe)
        print('')

    stimframes_df.to_csv(os.path.join(data_path, 'stimframes.csv'))
    print('Compiled stimframes')
    print('')
```
