```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from scipy import interpolate
```

```python
input_path = r'/Volumes/Behaviour/Tailored 3sec 2P Trials/PCA'
output_path = r'/Volumes/Behaviour/Tailored 3sec 2P Trials/PCA/PCA UpSampled'
sessions = [session for session in os.listdir(input_path) if session.endswith('.csv')]
len(sessions)
```

```python
session_names = []
for session in sessions:
    if session.startswith('._'): # circumventing issues with invisible files on external drives
        continue
    
    print('processing: ', session)
    entries = session.split('_')
    
    if session.endswith('CFA.csv'): 
        ses_name = entries[3]+'_'+entries[4][:3]
        
    else:
        ses_name = entries[3]+'_'+entries[4][:5]
            
    print(ses_name)
    session_names.append(ses_name)
    
    pca = pd.read_csv(os.path.join(input_path,session), index_col=0)
    rois = list(pca.columns[:-1])
    
    trials = pca['identifier'].unique()
    
    PCA_UP = pd.DataFrame()
    for r, roi in enumerate(rois):
        pca_r = pca[[roi, 'identifier']]
        traces_up = pd.DataFrame()
        for t, trial in enumerate(trials):
            trace = pca_r[pca_r['identifier']==trial]
            # up-sample neural recording to beahviour frame rate (400fps)
            x = np.arange(1/6.1, 21/6.1, 1/6.1)
            y_samples = np.array(trace[roi])
            x = np.array(x)
            xnew = np.arange(1/6.1, 20/6.1, 1/400)
            f = interpolate.interp1d(x, y_samples,kind='linear')
            ynew = f(xnew)
            trace_up = pd.DataFrame(ynew[:-2])
            #print('up-sampled length: ', len(dT_T0_up))
            traces_up = pd.concat([traces_up, trace_up], axis=0)
            
        new_name = ses_name + '_' + str(roi)
        PCA_UP[new_name] = traces_up[0]
        
    # add identifier column to dF_F0 dataframes 
    ids = []
    for t in trials:
        for n in range(1244):
            ids.append(t)                                       
    ids = np.array(ids) 
    
    PCA_UP['identifier'] = ids
 
    PCA_UP.to_csv(os.path.join(output_path, 'pca_data_up_'+ses_name+'.csv'))
```
